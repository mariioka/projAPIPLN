```python
import re
from collections import Counter

df['Comentario'] = df['Comentario'].str.lower()

stopwords = [
    'a', 'o', 'as', 'os', 'um', 'uma', 'uns', 'umas', 'ao', 'à', 'aos', 'às', 'do', 'da', 'dos', 'das', 'no', 'na',
    'nos', 'nas', 'pelo', 'pela', 'pelos', 'pelas', 'eu', 'ele', 'ela', 'nós', 'eles', 'elas', 'me', 'nos', 'meu',
    'minha', 'meus', 'minhas', 'nosso', 'nossa', 'nossos', 'nossas', 'de', 'em', 'por', 'para', 'com', 'sem', 'sob',
    'sobre', 'entre', 'até', 'desde', 'fora', 'e', 'mas', 'ou', 'porque', 'pois', 'que', 'se', 'como', 'quando',
    'também', 'é', 'muito', 'não', 'bem','mais','estava', 'fila', 'são', 'pra', 'já', 'tem', 'foi', 'tudo',
    'sempre', 'pouco', 'você', 'lugar', 'ser', 'vale', 'fui', 'achei', 'pena', 'ir', 'ter', 'pedi', 'só', 'porém', 'quem',
    'comer', 'super', 'outra', 'nem', 'mesmo', 'assim', 'isso', 'sei',
]

todosComentarios = " ".join(df['Comentario'])
todosComentarios = re.sub(r'[^\w\s]', '', todosComentarios)
palavras = todosComentarios.split()
palavras_filtradas = [p for p in palavras if p not in stopwords]

Counter(palavras_filtradas).most_common(20)
```

```python
import matplotlib.pyplot as plt

top_palavras = Counter(palavras_filtradas).most_common(20)

palavras = [p[0] for p in top_palavras]
frequencias = [p[1] for p in top_palavras]

plt.bar(palavras, frequencias)
plt.title("Palavras mais frequentes nas avaliações")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import re

def top_palavras_por_nota(df, nota, stopwords, n=10):
    df_filtrado = df[df['Nota'] == nota]
    texto = " ".join(df_filtrado['Comentario'].dropna())
    texto = texto.lower()
    texto = re.sub(r'[^\w\s]', '', texto)
    palavras = texto.split()
    palavras_filtradas = [p for p in palavras if p not in stopwords]
    return Counter(palavras_filtradas).most_common(n)

for nota in sorted(df['Nota'].unique()):
    top_palavras_nota = top_palavras_por_nota(df, nota, stopwords, n=10)
    palavras = [p[0] for p in top_palavras_nota]
    freq = [p[1] for p in top_palavras_nota]

    plt.figure(figsize=(8,5))
    plt.bar(palavras, freq, color='skyblue')
    plt.title(f"Top palavras - Nota {nota}")
    plt.xticks(rotation=45)
    plt.ylabel("Frequência")
    plt.show()
```

```python
import re

def converter_data(texto):
    texto = str(texto).lower()

    texto = texto.replace('um mês', '1 mês')
    texto = texto.replace('uma semana', '1 semana')
    texto = texto.replace('um ano', '1 ano')

    numero = re.search(r'\d+', texto)

    if not numero:
        return None

    numero = int(numero.group())

    if 'semana' in texto:
        return round(numero / 4, 2)

    elif 'mês' in texto or 'meses' in texto:
        return numero

    elif 'ano' in texto:
        return numero * 12

    else:
        return None
```

```python
df['meses_atras'] = df['HaQuantoTempo'].apply(converter_data)
```

```python
df.groupby('meses_atras')['Nota'].mean()
```

```python
media_tempo = df.groupby('meses_atras')['Nota'].mean()

media_tempo = media_tempo.sort_index()

plt.figure(figsize=(8,5))

plt.plot(media_tempo.index, media_tempo.values, marker='o')

plt.xlabel("Meses atrás")
plt.ylabel("Nota média")
plt.title("Evolução das avaliações ao longo do tempo")

plt.ylim(3, 5.25)

plt.grid(alpha=0.3)

plt.show()
```
